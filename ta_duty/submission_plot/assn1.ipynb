```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
fig = plt.figure()
fig.suptitle('Assignment 1 submission stats', fontsize=20)
height = [1,0,0,0,0,1,14,22]
bars = ('12', '13', '14', '15', '16', '17', '18', '19')
y_pos = np.arange(len(bars))
plt.bar(y_pos, height)
plt.xticks(y_pos, bars)
plt.xlabel("Submission date of students", fontsize=12)
plt.ylabel('Number of students', fontsize=12)
plt.show()
print ("Not submitted : 3")
print ("Submissions received between 18th night and 19th morning : 36 (14+22)")
print ("Single submission on 12th and 17th")
```
